```python
# Importando libreras
import scipy.io as sio  # Scipy para abrir los archivos .mat
from scipy.io import loadmat
import numpy as np  # Numpy: librera para trabajar con matrices y operaciones numricas
import matplotlib.pyplot as plt  # Para graficar las seales
import pandas as pd  # Para trabajar con el DataFrame
import seaborn as sb

class Archivo_Mat:
    def __init__(self, file_name, key):
        self.__file_name = file_name
        self.__key = key
        self.__file = sio.loadmat(self.__file_name)[self.__key]
    def convertir_a_2D(self, matriz):
        matriz.reshape[matriz.shape[0], -1]
        return matriz
    
    
class Archivo_CSV:
    def __init__(self, file_name):
        self.__file_name = file_name
        self.__file = pd.read_csv(self.__file_name)
    def grafico_torta(self, column='Gender'):
        counts = self.__file[column].value_counts()
        plt.pie(counts, labels=["Masculino", "Femenino", "No aplica"], autopct='%1.1f%%')
        plt.title("Gneros de pacientes con Cncer")
        plt.show()
    def grafico_barras(self):
        sb.countplot(data=self.__file,x='Level',  hue='Gender')
        plt.title('Cantidad de pacientes por nivel de gravedad')
        plt.xlabel('Nivel de gravedad')
        plt.ylabel('Cantidad de pacientes')
        plt.show()
    def get_file(self):
        return self.__file


fm = Archivo_Mat(file_name='S1.mat', key='data')


fcsv = Archivo_CSV(file_name='cancer patient data sets.csv')
fcsv.grafico_torta()
fcsv.grafico_barras()


```
