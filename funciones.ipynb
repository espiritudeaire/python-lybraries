```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def f_derivate(x):
    return 2*x

def tangent_line(x, x0, y0):
    return slope * (x-x0)+y0



x = np.linspace(-5, 5, 1000)
y = f(x)

x0 = 3
y0 = f(x0)
slope = f_derivate(x0)

# Crear la grfica
#plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^2$', color='blue')
plt.plot(x, tangent_line(x, x0, y0), label='Linea tangente', color='red', ls='--')
plt.scatter(x0, y0)
plt.text(x0, y0, f'  ({x0}, {y0})', fontsize=10, verticalalignment='bottom')  # Etiqueta del punto de tangencia
plt.title('Grfica de $f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
