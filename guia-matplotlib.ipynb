```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Grfico simple')
plt.show()

fig, axs = plt.subplots(2 ,2) #crea figura con cuadrcula 2x2
fig.subplots_adjust(hspace=0.5, wspace=0.5) #aumenta espacio entre filas hspace y columnas wspace
axs[0,0].plot([1,2,3],[1,2,3], color='red', linewidth=2, linestyle='--')
axs[0,1].plot([1,2,3,4],[1,4,3,7])
axs[1,0].plot([1,4,7,100],[1000,1,34,5])
axs[1,1].plot([1,2,3,4,54],[1,2,3,4,60])

plt.show

fig, axs = plt.subplot_mosaic([['left','right_top'], ['left', 'right_bottom']])
axs['left'].plot([1,2,3,4],[1,4,3,7], label='Izquierda')
axs['right_top'].plot([1,4,7,100],[1000,1,34,5], label='Derecha Arriba')
axs['right_bottom'].plot([1,2,3,4,54],[1,2,3,4,60], label='Derecha abajo')
plt.show()

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].plot([1, 2, 3], [3, 2, 1])
axs[1, 0].plot([1, 2, 3], [2, 3, 1])
axs[1, 1].plot([1, 2, 3], [1, 3, 2])
plt.show()

data = {'a': np.arange(50), 'b': np.random.randn(50)*10}
fig, ax = plt.subplots()
ax.scatter('a','b', data=data)
ax.set_xlabel('Variable A')
ax.set_ylabel('Variable B')
plt.show()

def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})

plt.show()
```
