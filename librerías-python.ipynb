```python
import pandas as pd  # Importamos la librería

def salto_linea(elemento=""):
    print(elemento)

# Creamos un diccionario con los datos
data = {
    'Nombre': ['Ana', 'Luis', 'Pedro'],
    'Edad': [28, 22, 35],
    'City': ['Madrid', 'Barcelona', 'Sevilla']
}

# Creamos el DataFrame
df = pd.DataFrame(data)

# Mostramos el DataFrame
print(df)
print(df["Edad"])
print(df.loc[2].Nombre)

productos = {
    'product': ['libra de arroz', 'galletas', "pan bimbo", "Azucar blanca", "azucar morena"],
    'price': [5000, 6000, 2000, 3200, 4300],
    'stock': [56, 17, 80, 101, 3]
}

tabla_productos = pd.DataFrame(productos)
print(tabla_productos)
print(tabla_productos["product"])
print(tabla_productos.loc[0])
salto_linea("----------")
#print(tabla_productos.product)
print(tabla_productos[['product', 'price']])
salto_linea("----------")

tabla_productos['total_value'] = tabla_productos['price'] * tabla_productos['stock']
tabla_productos.sort_values(by = 'price', ascending=True)
print(tabla_productos)
salto_linea()

#filtrando
filtro1 = tabla_productos[tabla_productos['price']>4000]



salto_linea('FILTRO 1----------------')
print(filtro1)

filtro2 = tabla_productos[tabla_productos['stock']>50]
salto_linea('FILTRO 2----------------')
print(filtro2)

filtro3 = tabla_productos[tabla_productos['product'].str.contains('azucar', case=False)]
salto_linea('FILTRO 3----------------')
print(filtro3)

productos_bajo_stock = tabla_productos[tabla_productos['stock']<20]
salto_linea('BAJO STOCK----------------')
print(productos_bajo_stock)

tabla_productos.to_csv('productos.csv', index=False)
tabla_leida = pd.read_csv('productos.csv')
salto_linea('TABLA LEÍDA---------------')
print(tabla_leida)
```
